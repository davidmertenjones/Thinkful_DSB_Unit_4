```python
top_20_authors = df['author'].value_counts().head(20).index
```

```python
top_20_authors
```

```python
articlelengths = []

for article in df[df['author'].isin(top_20_authors)]['text'].values:
    articlelengths.append(len(article))
```

```python
# Number of articles written by top 20 authors:

len(articlelengths)
```

```python
# Total character length of included articles before text cleaning:

sum(articlelengths)
```
